```python
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import rasterio
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error, accuracy_score, r2_score
import cv2
import gdal
from rasterio.transform import from_origin
from rasterio.mask import mask
import geopandas as gpd
```

```python
LST_1km = rasterio.open(r'H:\ADM-Kenya\Results\LST\Busia\Busia_20210615_2.tif')
NDVI_1km = rasterio.open(r'H:\ADM-Kenya\Results\LST\Busia\Busia_20210615_3.tif')
NDVI_10m = rasterio.open(r'H:\ADM-Kenya\Results\LST\Busia\mNDVI_Image_2021-06-10_clipped.tif')
```

```python
dataset = gdal.Open(r'H:\ADM-Kenya\Results\LST\Busia\Busia_20210615_2.tif')
bands = dataset.GetRasterBand(1)

print('Number of Bands = ', dataset.RasterCount, '\nNumber of Horizontal Pixels = ', dataset.RasterXSize,  
      '\nNumber of Vertical Pixels = ', dataset.RasterYSize, '\nBands: ', bands)
```

```python
# Read the actual data arrays
lst_1km = LST_1km.read(1)
ndvi_1km = NDVI_1km.read(1)
ndvi_10m = NDVI_10m.read(1)
```

```python
ndvi_upscaled = cv2.resize(ndvi_10m, (lst_1km.shape[1], lst_1km.shape[0]), interpolation=cv2.INTER_NEAREST)
```

```python
lst_1km = lst_1km.astype(float)
lst_1km[lst_1km == -9999] = np.nan
np.nanmean(lst_1km)
```

```python
plt.imshow(lst_1km, cmap='hot')
plt.colorbar(label='Temp')
plt.title('1- km LST Map')
plt.show()
```

```python
X = ndvi_upscaled.flatten().reshape(-1, 1)
y = lst_1km.flatten()
X.shape, y.shape
```

```python
# Mask out NaN values
valid_mask = ~np.logical_or(np.isnan(X[:, 0]), np.isnan(y))
X_valid = X[valid_mask]
y_valid = y[valid_mask]
X_valid.shape, y_valid.shape, np.mean(y_valid)
```

```python
regression_model_rf = RandomForestRegressor(n_jobs=-1)
regression_model_rf.fit(X_valid, y_valid)
```

```python
predictions_1km_rf = regression_model_rf.predict(X_valid)
residuals_rf = y_valid - predictions_1km_rf
#residuals_rf
```

```python
ndvi_upscaled.shape, lst_1km.shape, np.nanmax(lst_1km)
```

```python
X_10m = ndvi_10m.flatten().reshape(-1, 1)

# Mask out NaN values in the 10 m NDVI
valid_mask_10m = ~np.isnan(X_10m[:, 0])
X_valid_10m = X_10m[valid_mask_10m]
```

```python
downscaled_lst_10m = regression_model_rf.predict(X_valid_10m)
```

```python
downscaled_lst_nan = np.empty_like(ndvi_10m)
downscaled_lst_nan[:] = np.nan
downscaled_lst_nan[valid_mask_10m.reshape(ndvi_10m.shape)] = downscaled_lst_10m
print("Shape of downscaled LST:", downscaled_lst_nan.shape)
```

```python
plt.imshow(downscaled_lst_nan, cmap='hot')
plt.colorbar(label='Temperature')
plt.title('Downscaled LST Map - 10m')
plt.show()
```

```python
# Define the output file path
output_file = r'H:\ADM-Kenya\Results\LST\Busia\LST_Busia_10m.tif'

# Get the metadata from the NDVI raster
ndvi_metadata = NDVI_10m.meta

# Update the metadata for the output GeoTIFF
dst_metadata = {
    'driver': 'GTiff',
    'dtype': downscaled_lst_nan.dtype,
    'nodata': np.nan,
    'width': downscaled_lst_nan.shape[1],
    'height': downscaled_lst_nan.shape[0],
    'count': 1,  # Assuming a single-band GeoTIFF
    'crs': ndvi_metadata['crs'],
    'transform': ndvi_metadata['transform']
}

# Save the array as a GeoTIFF
with rasterio.open(output_file, 'w', **dst_metadata) as dst:
    dst.write(downscaled_lst_nan, 1)  # Assuming a single-band GeoTIFF
```

```python
# Create an array with NaN values of the same shape as ndvi_10m
res_lst_nan = np.empty_like(lst_1km)
res_lst_nan[:] = np.nan
res_lst_nan[valid_mask.reshape(lst_1km.shape)] = residuals_rf
```

```python
plt.imshow(res_lst_nan, cmap='hot')
plt.colorbar(label='Temp')
plt.title('1-km residuals Map')
plt.show()
```

```python
res_dwscaled = cv2.resize(res_lst_nan, (ndvi_10m.shape[1], ndvi_10m.shape[0]), interpolation=cv2.INTER_NEAREST)
res_dwscaled.shape
```

```python
lst_c_10m = downscaled_lst_nan + res_dwscaled
```

```python
plt.imshow(lst_c_10m, cmap='hot')
plt.colorbar(label='Temp')
plt.title('10-m corrected Map')
plt.show()
```

```python
plt.scatter(X_valid, residuals_rf)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('1 km NDVI')
plt.ylabel('Residuals')
plt.title('Residual Plot Rnadom Forest')
plt.show()
```
